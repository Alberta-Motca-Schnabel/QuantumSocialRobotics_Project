```python
from datasets import load_dataset
import pandas as pd
#import nltk
from Utility import clean_dataset

#nltk.download('punkt')
#nltk.download('stopwords')

ds = load_dataset("google-research-datasets/go_emotions", "simplified")
ds_train = ds['train'].to_pandas()
ds_test = ds['test'].to_pandas()
ds = pd.concat([ds_train, ds_test])
print(ds.head())
print(ds.shape)
```

```python
emotion = { 0 : 'admiration',
            1 : 'amusement',
            2 : 'anger',
            3 : 'annoyance' ,
            4 : 'approval' ,
            5 : 'caring' ,
            6 : 'confusion',
            7 : 'curiosity',
            8 : 'desire',
            9 : 'disappointment',
            10: 'disapproval' ,
            11: 'disgust',
            12: 'embarrassment',
            13: 'excitement',
            14: 'fear',
            15: 'gratitude',
            16: 'grief',
            17: 'joy',
            18: 'love',
            19: 'nervousness',
            20: 'optimism',
            21: 'pride',
            22: 'realization',
            23: 'relief',
            24: 'remorse',
            25: 'sadness',
            26: 'surprise',
            27: 'neutral'}

#Ekman representation
ekman_emotion = { "angry": ["anger", "annoyance", "disapproval"],
                  "disgust": ["disgust"],
                  "fear": ["fear", "nervousness"],
                  "happy": ["joy", "amusement", "approval", "excitement", "gratitude",  "love", "optimism", "relief", "pride", "admiration", "desire", "caring"],
                  "sad": ["sadness", "disappointment", "embarrassment", "grief",  "remorse"],
                  "surprise": ["surprise", "realization", "confusion", "curiosity"],
                  "neutral": ["neutral"]
}

label_proccesed = []
for label in ds['labels']:
    label_proccesed.append(emotion[int(label[0])])

labels = []
for label in label_proccesed:
    for key, value in ekman_emotion.items():
        if label in value:
            labels.append(key)

ds['labels'] = labels
ds
```

```python
ds = ds.drop('id', axis=1)
ds
```

```python
texts = clean_dataset(ds)
ds['text'] = texts
ds
```
